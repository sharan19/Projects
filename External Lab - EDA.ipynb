##                                          HR DATA SET

Steps - 
1. Import the necessary libraries
2. Load the dataset and explore it.
3. Count the male and female employees and plot it.
4. Count the employee attrition and plot it.
5. Make a stacked barplot for employee attrition based on gender.

```python
import pandas as pd
import numpy as np
import matplotlib as plt
import csv
```

```python
hr=pd.read_csv('HR.txt',delim_whitespace=True)
```

```python
hr.head()
```

```python
# Printing the number of rows and columns in the dataset
print("Total number of rows in HR dataset are   :",hr.shape[0])
print("Total number of columns in HR dataset are:",hr.shape[1])
```

```python
#Finding datatypes 
hr.dtypes
```

```python
#counting number of Attrition 
hr.groupby(['Attrition']).size()
```

```python


hr.loc[0]
```

```python
 df1=hr.loc[lambda hr: hr['Gender'] == 1, :]
print("Male Employees average monthly salary:")    
df1.loc[:,'MonthlyIncome'].mean()
```

```python
 df1=hr.loc[lambda hr: hr['Gender'] == 2, :]
print("Female Employees average monthly salary:")    
df1.loc[:,'MonthlyIncome'].mean()
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline
```
